## Carga de datos
Se realiza carga de datos obtenidos mediante web crawling a pollstar.com

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
from pandas.plotting import lag_plot
import numpy  as np
```

```python
filename = "/home/bberrios/experimento-1/filtrar/live_boxoffice-all-4.txt"
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y')
events = pd.read_csv(filename, sep='\t', parse_dates=['StartPlayDate'], date_parser=dateparse)
events.head()
```

```python
print 'cantidad de datos: ', len(events)
```

## ¿Qué artista es el más exitoso?

```python
successs_threhold = 80
```

```python
success_events = events[events.capacityPercentage >= successs_threhold]

grouping = success_events.groupby('sortName', as_index = False)
df = grouping['capacityPercentage'].agg(['mean','std', 'median', 'count']).replace(np.nan, 0, regex=True)
df=df.reset_index()
```

### Con más de 1 evento

```python
succes_gt_1 = df[df['count'] > 1].sort_values(by = ['mean', 'count'], ascending = [0, 0])
print "Cantidad: ", len(succes_gt_1)
succes_gt_1
```

### Con más de 5 eventos

```python
succes_gt_5 = df[df['count'] > 5].sort_values(by = ['mean', 'count'], ascending = [0, 0])
print "Cantidad: ", len(succes_gt_5)
succes_gt_5
```

### Con más de 10 eventos

```python
succes_gt_10 = df[df['count'] > 10].sort_values(by = ['mean', 'count'], ascending = [0, 0])
print "Cantidad: ", len(succes_gt_10)
succes_gt_10
```
